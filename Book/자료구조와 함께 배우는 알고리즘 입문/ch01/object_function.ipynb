```python
# 실습 1C-4
# 함수 내부, 외부에서 정의한 변수와 객체의 식별 번호를 출력하기
# n과 x 모두 int형 객체 1을 참조하는 이름에 불과

n = 1          # 전역 변수(함수 내부, 외부에서 사용)
def put_id():
    x = 1      # 지역 변수(함수 내부에서만 사용)
    print(f'id(x) = {id(x)}')

print(f'id(1) = {id(1)}')
print(f'id(n) = {id(n)}')
put_id()
```

```python
# 실습 1C-5
# 1부터 100까지 반복하여 출력하기
for i in range(1, 101):
    print(f'i = {i:3}   id(i) = {id(i)}')  # {i:3}백의자리(100) 출력을 위해 3칸 띄우기
```
